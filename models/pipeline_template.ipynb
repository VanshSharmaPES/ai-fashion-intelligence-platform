### Step 2: Initialize CLIP Model

```python
import torch
from transformers import CLIPProcessor, CLIPModel
from PIL import Image

# Load Model
model_id = "openai/clip-vit-base-patch32"
processor = CLIPProcessor.from_pretrained(model_id)
model = CLIPModel.from_pretrained(model_id)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
print(f"CLIP Model loaded on {device}!")
```

### Step 3: Example - Embedding a Vibe (Text)
Pass the user's abstract query to extract a text embedding.

```python
query = ["Rainy day cozy cafe vibe sweater"]
inputs = processor(text=query, return_tensors="pt", padding=True).to(device)

with torch.no_grad():
    text_embeddings = model.get_text_features(**inputs)
    text_embeddings = text_embeddings / text_embeddings.norm(p=2, dim=-1, keepdim=True) # Normalize

print("Query Vector Shape:", text_embeddings.shape)
```
